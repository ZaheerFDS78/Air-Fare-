#EDA

from description we can see Date_of_Journey is a object data type.
Therefore, we have to convert this datatype into timestamp so as to use this column properly for prediction 

for this we require pandas to_datetime to convert object datatype to datetime dtype

.dt.day method will extract only day of that date/

```python
df["Journey_Day"]=pd.to_datetime(df["Date_of_Journey"],format='%d/%m/%Y').dt.day
```

```python
df["Journey_Month"]=pd.to_datetime(df["Date_of_Journey"],format='%d/%m/%Y').dt.month
```

```python
df.head()
```

```python
#Since Date of Journey is no more required we extracted the Day and Month of Journey in seprate columns
df.drop(["Date_of_Journey"], axis=1, inplace=True)
```

```python
# Departue time is the time where Plane leaves the Current Airport 
# We can also extract the Hour and Min with the same datetime method

df["Dep_Hour"]=pd.to_datetime(df["Dep_Time"]).dt.hour
df["Dep_Minute"]=pd.to_datetime(df["Dep_Time"]).dt.minute
```

```python
df.head()
```

```python
#Now we dont need the Dep_Time because we extracted the value from it 
df.drop(["Dep_Time"],axis=1, inplace=True)
```

```python
df.head()
```

```python
#Now same process goes for the Arrival time as well we will extract the values and
#delete the Dep_Time Colum

df["Arrival_Hour"]=pd.to_datetime(df["Arrival_Time"]).dt.hour
df["Arrival_Min"]=pd.to_datetime(df["Arrival_Time"]).dt.minute
```

```python
df.head()
```

```python
df.drop(["Arrival_Time"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df[['Duration_Hour','Duration_Minute']] = df.Duration.apply(
   lambda x: pd.Series(str(x).split(" ")))
```

```python
df.head()
```

```python
#replacing the string values in both the columns afater spliting
# Removing the h from Duration Hour column
# Re 
df['Duration_Hour'] = df.Duration_Hour.str.replace('h' , '')
df['Duration_Minute'] = df.Duration_Minute.str.replace('m' , '')
```

```python
df.head(20)
```

```python
#Checking with the all the null values in the dataframe
df.isnull().sum()
```

```python
#eplacing the null values in the duration minute column
df["Duration_Minute"]=df["Duration_Minute"].fillna(0)
```

```python
#Couting and checing the Null values in dataframe
df.isnull().sum()
```

```python
df.drop(['Duration'], axis=1, inplace=True)
```

```python
# As Source is also Nominal Categorical Data we will perform OneHotEncoding
Source=df['Source']
Source=pd.get_dummies(Source, drop_first=True)
Source.head()
```

```python
# on the destination feature the OneHOtEncoding 

Destination=df["Destination"]
Destination=pd.get_dummies(Destination, drop_first=True)
Destination.head()
```
